```python
# OpenCV Lab 1: Drawing Functions
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a blank 512x512 image
img = np.zeros((512, 512, 3), dtype=np.uint8)

# Draw a red line
cv2.line(img, (50, 50), (200, 200), (0, 0, 255), 2)

# Draw a green rectangle
cv2.rectangle(img, (100, 300), (300, 400), (0, 255, 0), 3)

# Draw a blue filled circle
cv2.circle(img, (400, 100), 50, (255, 0, 0), -1)

# Writing on Images
cv2.putText(img, 'MUHAMMAD FAIZAN', (210, 230), cv2.FONT_HERSHEY_SIMPLEX, 1, (100, 210, 255), 3)


# Display and save
plt.imshow(img)
```
